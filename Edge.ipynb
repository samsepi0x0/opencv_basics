```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('rena.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

canny = cv2.Canny(gray, 150, 200)
cv2_imshow(canny)

# EROSION AND DILUTION:
# Erosion is used to remove the noise level from the image
kernel = np.ones((1, 1), np.uint8) # kernel is just the window which erodes the image
erode = cv2.erode(canny, kernel, iterations=1)
cv2_imshow(erode)
```

```python
#DiLATIoN
kernel = np.ones((5, 5), np.uint8)
cv2_imshow(canny)
dilated = cv2.dilate(canny, kernel, iterations=1)
cv2_imshow(dilated)
```

```python
## IF WE USE EROSION AFTER DILATION:
cv2_imshow(canny)
erode2 = cv2.erode(dilated, kernel, iterations=1)
cv2_imshow(erode2)
```

```python
#Display images horizontally
display = np.hstack((canny, dilated, erode2))
cv2_imshow(display)
```
